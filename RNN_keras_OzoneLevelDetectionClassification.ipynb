```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.layers import *
from keras.models import *
from keras.utils import *
from sklearn.preprocessing import *
```

```python
label = ['feat_{}'.format(i) for i in range(73)]

df = pd.read_csv('eighthr.data', names=label)
df
```

```python
df.dtypes

df = df.apply(pd.to_numeric, errors='coerce') # ë°ì´í„°í”„ë ˆìž„ì˜ ê°’ë“¤ì„ ìˆ«ìžë¡œ ë°”ê¿€ ìˆ˜ ì—†ëŠ” ê°’ë“¤ì„ NaNìœ¼ë¡œ ë°”ê¿”ì£¼ëŠ”ê±°
df
```

```python
# df.dtypes
df.dropna(inplace=True) #NaNì´ë¼ëŠ” ê°’ì„ ë‹¤ ë²„ë¦¬ëŠ” ê²ƒ, ëŒ€ì‹ ì— ë°ì´í„° ì‚¬ì´ì¦ˆê°€ ì¤„ì–´ë“¬
df
```

```python
# df.describe()
Y = df['feat_72']
Y
```

```python
Y = to_categorical(Y)
```

```python
df.drop(["feat_72"], axis=1, inplace=True)

df
```

```python
X_train = np.asarray(df[:-100].values.tolist(), dtype=np.float64)
X_test = np.asarray(df[-100:].values.tolist(), dtype=np.float64)

Y_train = Y[:-100]
Y_test = Y[-100:]
```

```python
X_train = X_train[:1700]
Y_train = Y_train[:1700]
```

```python
#(sample, timestep, feature) 3ì°¨ì› ê°’ì„ ì¤˜ì•¼í•¨
X_train = X_train.reshape(-1, 10, 72)
```

```python
Y_train = Y_train.reshape(-1,10, 2)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
#many to many
model = Sequential()
model.add(LSTM(128, input_shape=(10, 72), return_sequences=True)) #return_sequencesëŠ” ìœ„ì— ê°’ì„ ë‹¤ì‹œ ë”°ì˜¤ëŠ”ê±° ì—†ìœ¼ë©´ many to one
model.add(LSTM(256, return_sequences=True))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train, Y_train, epochs=10, batch_size=1, validation_split=0.1)
```

```python
X_test = X_test.reshape(-1, 10, 72)
Y_test = Y_test.reshape(-1, 10, 2)
```

```python
score = model.evaluate(X_test, Y_test)
print(score)
```
