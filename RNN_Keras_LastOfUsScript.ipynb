```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.layers import *
from keras.models import *
from keras.utils import *
from sklearn.preprocessing import *
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import random
import re
```

```python
file = 'the_last_of_us.txt'

text = ''
array = []
with open(file, encoding="utf-8") as f:
    lines = f.readlines()
    
    text = text.join([l for l in lines if re.match(r'^[A-Z].*:', l)]) #      
    # [A-Z]   :          
    
for t in text.split('\n')[:1000]: #  
    array.append(t) #array  

array
```

```python
token = Tokenizer(lower=False, filters='.,?;\'\"-')
token.fit_on_texts(array)
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
seq = token.texts_to_sequences(array)
seq

#(sample, timestep, feature)
# I am student
```

```python
seq = pad_sequences(seq, maxlen=10)
seq
```

```python
X = seq
```

```python
Y = np.vstack((X[1:], X[0])) # Y X        
Y
```

```python
Y.shape
```

```python
X = X.reshape(-1, 10, 1)
Y = Y.reshape(-1, 10, 1)
X
```

```python
X.shape
```

```python
Y = to_categorical(Y)
Y
```

```python
Y.shape
```

```python
'''
Bidirectional 
-> -> -> ->
<- <- <- <-
GRU/LSTM
'''

model = Sequential()
model.add(Bidirectional(SimpleRNN(128, return_sequences=True), input_shape=(10, 1)))
model.add(Bidirectional(SimpleRNN(128, return_sequences=True)))
model.add(Dense(1385, activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs=100, batch_size=1)
model.save('model.save')
```
